# Specific Communication between CT1 and CT2

```python
ct1_indices = subset.obs[subset.obs['cellType'] == 'Inflammatory.macrophages'].index
ct2_indices = subset.obs[subset.obs['cellType'] == 'CD3+.alpha.beta.T.cells'].index

ligand_receptor_pairs = subset.uns['selected_spots']

active_interactions_ct1 = ligand_receptor_pairs[ct1_indices]
active_interactions_ct2 = ligand_receptor_pairs[ct2_indices]
```

```python
active_interactions_ct1
```

```python
# Convert indices to positions in the matrix
# don't use np.searchsorted as suset.obs.index is not sorted
ct1_positions = [np.where(subset.obs.index == x)[0][0] for x in ct1_indices if x in subset.obs.index]
ct2_positions = [np.where(subset.obs.index == x)[0][0] for x in ct2_indices if x in subset.obs.index]

# Initialize lists to store filtered indices
filtered_ct1_indices = []
filtered_ct2_indices = []
communication_pairs = []

# Iterate over pairs of indices
for i in ct1_positions:
    for j in ct2_positions:
        # Check if the weight between the two cells is non-zero
        if subset.obsp['weight'][i, j] > 0:
            # If yes, add the indices to the filtered lists
            filtered_ct1_indices.append(subset.obs.index[i])
            filtered_ct2_indices.append(subset.obs.index[j])
            communication_pairs.append([subset.obs.index[i],subset.obs.index[j]])

# Convert lists back to pandas index
filtered_ct1_indices = pd.Index(filtered_ct1_indices)
filtered_ct2_indices = pd.Index(filtered_ct2_indices)
```

```python
# Initialize a dictionary to store counts of common interactions
common_interactions_count = {}

for comm_pair in communication_pairs:
    # Get active interactions for each cell in the pair
    LR1 = active_interactions_ct1[comm_pair[0]]
    LR2 = active_interactions_ct2[comm_pair[1]]

    # Find common active interactions
    common_interactions = LR1.index[LR1 & LR2]

    # Update counts in the dictionary
    for interaction in common_interactions:
        if interaction in common_interactions_count:
            common_interactions_count[interaction] += 1
        else:
            common_interactions_count[interaction] = 1
            
# Sort the dictionary by values (from most to least)
sorted_common_interactions = sorted(common_interactions_count.items(), key=lambda x: x[1], reverse=True)
```

```python
len(communication_pairs)
```

```python
# Convert sorted_common_interactions to a DataFrame
LRI_df = pd.DataFrame(sorted_common_interactions, columns=['Interaction', 'Count'])

# Calculate the total count for percentage calculation
total_count = LRI_df['Count'].sum()

# Add a new column for percentage
LRI_df['Percentage'] = (LRI_df['Count'] / total_count) * 100

# Sort the DataFrame by 'Count' or 'Percentage' and select the top 10
df_top10 = LRI_df.sort_values(by='Count', ascending=False).head(10)

# Create the barplot for the top 10 items
plt.figure(figsize=(6, 6))  # Adjust the size as needed
ax = sns.barplot(x='Percentage', y='Interaction', data=df_top10)

# Reduce the font size for y-tick labels
plt.yticks(fontsize=8)  # Adjust the font size as needed

# Add percentage at the top edge of each bar
for p in ax.patches:
    ax.text(p.get_width(), p.get_y() + p.get_height() / 2, 
            '{:.1f}%'.format(p.get_width()),  # Format the text to show one decimal place
            fontsize=8,  # Adjust the font size as needed
            va='center')

# Add titles and labels
plt.title('Top 10 Common LRIs')
plt.xlabel('Percentage')
plt.ylabel('Interaction')
plt.tight_layout()

# Save the plot
plt.savefig('../figures/liver/'+test_sample+'_LRIs_percentages.jpeg')
plt.show()
```
